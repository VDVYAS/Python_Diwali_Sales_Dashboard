```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape')
```

```python
df.head()
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(7,3)})
df["Gender"].value_counts(ascending=False).plot.pie(autopct='%1.0f%%')
plt.title('Gender',fontsize=20)
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum()
sales_gen
width= 0.2
sns.barplot(x="Gender",y="Amount",data=sales_gen,width=width)
```

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(10,3)})
sales=sns.countplot(data=df, x= "Age Group",hue="Gender",order=df['Age Group'].value_counts(ascending=True).index)
plt.title("Selling by Age Group")

for bars in sales.containers:
    sales.bar_label(bars)
```

Above Graph we can say that Female in buying more than Male. and specially age Group 26-35 women is buying

```python
sales_gen=df.groupby(['Age Group'], as_index=False)['Amount'].sum()
ax=sns.barplot(x="Age Group",y="Amount",data=sales_gen,width=0.4,order=df['Age Group'].value_counts(ascending=True).index)
```

From above graph we can say that highest purchasing power is group 26-35

# State


```python
sns.set(rc={'figure.figsize':(7,3)})
df["State"].value_counts(ascending=False).head().plot.pie(autopct='%1.0f%%')
plt.title('State',fontsize=20)
```

From above graph we can say that most of order come from uttar pradesh,Maharshtra and Karnataka.

```python
sales_gen=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sales_gen
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data=sales_gen,x='State',y="Amount")
plt.xlabel('State',fontsize=20)
plt.ylabel('Amount',fontsize=20)
```

```python
sns.set(rc={'figure.figsize':(7,3)})
df["Marital_Status"].value_counts().head().plot.pie(autopct='%1.0f%%')
plt.title('Marital Status',fontsize=20)
```

From above graph we can say that marrige couple are purchasing more that single

```python
sales_gen=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,3)})
ax=sns.barplot(x="Marital_Status",y="Amount",data=sales_gen,width=0.4,)
```

From above grapht we scan say that highest purchasing power is married

```python
sales_gen=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.barplot(data=sales_gen,x="Occupation",y="Amount")
plt.xlabel('Occupation',fontsize=20)
plt.ylabel('Amount',fontsize=20)
```

From above graph we can say that most of the buyer is IT Sector, Heathcare and Aviation.

```python
df["Product_Category"].value_counts().head().plot.pie(autopct='%1.0f%%')
plt.title('Product category',fontsize=20)
```

From above graph we can say that most of that purchasing is clothin & apparel, Food and Eletronics & Gadgets

```python
sns.set(rc={'figure.figsize':(20,5)})
sales_gen=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
ax=sns.barplot(x="Product_Category",y="Amount",data=sales_gen)
plt.xlabel('Product category',fontsize=20)
plt.ylabel('Amount',fontsize=20)
```

From above graph we can say that most sold product are Food,clothing & Apperal and Eletronics category

# Top Selling Product
